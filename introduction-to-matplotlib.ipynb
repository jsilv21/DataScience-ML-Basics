## Matplotlib example workflow


```python
#0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt
#1. prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2. setup plot
fig, ax = plt.subplots(figsize=(10,10))
#figsize = hight & width

#3. Plot data
ax.plot(x,y)

#4. Customize plot
ax.set(title="Simple Plot", 
      xlabel="x-axis",
      ylabel="y-ais")

#5. Save & Show (you save the whole figure)
fig.savefig('images/sample-plot2.png')
```

### Two options for subplots 


```python
#subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                            ncols = 2,
                                            figsize=(10,5));
#plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# subplots option 2
fig, ax = plt.subplots(nrows=2, ncols = 2, figsize=(10,5))

#plot to each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

 ### which one should you use? PyPlot vs. Matplotlib Object Oriented (OO) Method?

* when plotting quickly, OK to use pyplot method
* when plotting advanced, use OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
len(over_50)
```

```python
over_50.head()
```

```python
#PyPlot Method
over_50.plot(kind ="scatter",
             x="age",
             y="chol",
             c="target");
```

```python
#OO Method Mixed with PyPlot Method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind ="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);
#ax.set_xlim([45,100])
```

```python
#OO from SCRATCH

fig, ax = plt.subplots(figsize=(10,6))

#plot data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])
#customize
ax.set(title="Over 50 - Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol")

#add legend
#star value grabs values from scatter, use the elements to create the legend
ax.legend(*scatter.legend_elements(), title="Target")

#add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='dashdot');
```

```python
over_50.head()
```

```python
#Subplot of chol, age, thalac (max hr)
#SHARE X = True allows 2 subplots to share x label

#single figure, two Axes
fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1, figsize=(10,10), sharex=True)

#Add data to ax 0 
scatter = ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])
#customize
ax0.set(title="Over 50 - Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")

#add legend
#star value grabs values from scatter, use the elements to create the legend
ax0.legend(*scatter.legend_elements(), title="Target")

#add a horizontal line
ax0.axhline(over_50["chol"].mean(),
            linestyle='--')

#replicate with thalach column

scatter2 = ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"])
#customize
ax1.set(title="Over 50 - Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate (Thalach)")

#add legend
#star value grabs values from scatter, use the elements to create the legend
ax1.legend(*scatter.legend_elements(), title="Target")

#add a horizontal line
ax1.axhline(over_50["thalach"].mean(),
            linestyle='--')

#title to entire figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
#set style
plt.style.use('seaborn-whitegrid')

#OO from SCRATCH
fig, ax = plt.subplots(figsize=(10,6))

#plot data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter") #changes color scheme
#customize
ax.set(title="Over 50 - Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol")

#add legend
#star value grabs values from scatter, use the elements to create the legend
ax.legend(*scatter.legend_elements(), title="Target")

#add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='dashdot');
```

```python
#customizing y & x axis limitations
#Subplot of chol, age, thalac (max hr)
#SHARE X = True allows 2 subplots to share x label

#single figure, two Axes
fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1, figsize=(10,10), sharex=True)

#Add data to ax 0 
scatter = ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter")
#customize
ax0.set(title="Over 50 - Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")

#change x axis limits
ax0.set_xlim([50,80])

#add legend
#star value grabs values from scatter, use the elements to create the legend
ax0.legend(*scatter.legend_elements(), title="Target")

#add a horizontal line
ax0.axhline(over_50["chol"].mean(),
            linestyle='--')

#replicate with thalach column

scatter2 = ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"],
                    cmap="winter")
#customize
ax1.set(title="Over 50 - Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate (Thalach)")
ax1.set_ylim([60,200])
#add legend
#star value grabs values from scatter, use the elements to create the legend
ax1.legend(*scatter.legend_elements(), title="Target")

#add a horizontal line
ax1.axhline(over_50["thalach"].mean(),
            linestyle='--')

#title to entire figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
